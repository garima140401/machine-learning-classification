```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
np.random.seed(42)
teachers = np.random.choice(['Manoj', 'Hitesh', 'Gopal'], size=100)
teaching_style = np.random.randint(1, 6, size=100)
students_feedback = np.random.randint(1, 11, size=100)
scores_improvement = np.random.uniform(5, 20, size=100).round(2)
attendance = np.random.randint(50, 100, size=100)
subject_expertise = np.random.randint(1, 6, size=100)
```

```python
# Target
best_teacher = [
    1 if (ts >= 4 and sf >= 8 and si > 15 and att > 85 and se >= 4) else 0
    for ts, sf, si, att, se in zip(teaching_style, students_feedback, scores_improvement, attendance, subject_expertise)
]

# Create DataFrame
data = pd.DataFrame({
    'Teacher': teachers,
    'Teaching_Style': teaching_style,
    'Students_Feedback': students_feedback,
    'Scores_Improvement': scores_improvement,
    'Attendance': attendance,
    'Subject_Expertise': subject_expertise,
    'Best_Teacher': best_teacher
})
```

```python
data = pd.get_dummies(data, columns=['Teacher'], drop_first=True)

X=data.drop('Best_Teacher',axis=1)
y=data['Best_Teacher']

X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
data.head()
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred =clf.predict(X_test)
y_pred
```

```python
print("Accuracy", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
